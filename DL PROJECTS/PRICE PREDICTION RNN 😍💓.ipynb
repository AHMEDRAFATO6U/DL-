### IMPORT LIBRARY 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import *
import matplotlib.pyplot as plt
```

### SPLIT DATA 

```python
x=df[['feature1','feature2']]
y=df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 42)
```

### SCALING DATA 

```python
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

### BUILD MODEL (RNN)

```python
model = Sequential()

model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 10, activation = 'relu'))

model.add(Dense(units = 1))
```

### COMPILE MODEL

```python
model.compile(optimizer = 'adam', loss = "MSE")
```

### FIT MODEL 

```python
model.fit(x_train, y_train, epochs = 200)
```

```python
model.history.history
```

```python
lose_=pd.DataFrame(model.history.history)
lose_
```

```python
lose_.plot()
```

### EVALUTEION 

```python
model.evaluate(x_test, y_test)
```

```python
model.evaluate(x_train, y_train)
```

### TEST MODEL 

```python
y_pred=model.predict(x_test)
```

```python
y_pred=pd.Series(y_pred.reshape(250,))
y_pred
```

```python
y_test = pd.DataFrame(y_test.values, columns = ['True Values'])
y_test.head()
```

```python
new_df = pd.concat([y_test, y_pred], axis = 1)
new_df.head()
```

```python
new_df.columns = ['Actual Price', 'Predicted Price']
new_df.head()
```

```python
plt.plot(new_df['Actual Price'], color = 'red', label = 'Actual Price')
plt.plot(new_df['Predicted Price'], color = 'blue', label = 'Predicted Price')
plt.legend()
plt.show()
```
